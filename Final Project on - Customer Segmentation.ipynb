```python
# Import Liabiary
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
X = df.iloc[:,[3,4]].values   #using iloc function we can display
                              #3 and 4 columns from array 
X
```

```python
from sklearn.cluster import KMeans  # Import K-means Liabilary
```

```python
wcss = []    #Initializing the list for values of wcss

#Using for loop for iterations from 1 to 10

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init ='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) 
    
    # intertia measures how well a dataset was clustered by k-means .
    # It is calculated by measuring the distances between each data points
```

```python
# using Elbow method we can find Optimal no of clusterd
sns.set()                    
plt.plot(range(1,11), wcss)
plt.title('The elbow method')         
plt.xlabel('No of clusters')         
plt.ylabel('WCSS Values')         
plt.show() 
```

```python
kmeans = KMeans(n_clusters=5, init ='k-means++',random_state=0)
Y =kmeans.fit_predict(X)
print(Y)

#fit_predict will train our model & it is also define the cluster value..
```

```python
plt.figure(figsize=(8,8))   
plt.scatter(X[Y==0,0],X[Y==0,1], s=80,c='green',label='Cluster 1')  #
plt.scatter(X[Y==1,0],X[Y==1,1], s=80,c='red',label='Cluster 2')
plt.scatter(X[Y==2,0],X[Y==2,1], s=80,c='yellow',label='Cluster 3')
plt.scatter(X[Y==3,0],X[Y==3,1], s=80,c='violet',label='Cluster 4')
plt.scatter(X[Y==4,0],X[Y==4,1], s=80,c='blue',label='Cluster 5')


plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100 ,c='magenta',
            label='Centroids')
plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
plt.scatter(X[...,0],X[...,1])  # using we plot 0 column and 1 columns
plt.xlabel("Total Income")
plt.ylabel("Spending Income")
plt.show()
```

```python
import scipy.cluster.hierarchy as sch      # Import liabilary for hierachical clustering
dendrogram = sch.dendrogram(sch.linkage(X,method="ward")) # we plot dendrogram#calculated distances and simmilarites between all objects
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
Y_hc=hc.fit_predict(X) # using we find 0,1,2,3,4 make 5 cluster 
Y_hc
```

```python
plt.scatter(X[Y_hc==0,0],X[Y_hc==0,1],label = 'Careless')  # we plot 5 cluster 
plt.scatter(X[Y_hc==1,0],X[Y_hc==1,1],label = 'Sensible')
plt.scatter(X[Y_hc==2,0],X[Y_hc==2,1],label = 'Target')
plt.scatter(X[Y_hc==3,0],X[Y_hc==3,1],label = 'Standard')
plt.scatter(X[Y_hc==4,0],X[Y_hc==4,1],label = 'Careful')
plt.legend()
plt.show()
```
